# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in a.select_dtypes(include=['object']).columns:
    a[column]=le.fit_transform(a[column])
a
```

```python
a.corr()
```

```python
b=a.drop(["Country","Category"],axis="columns")
b
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
c=b.drop("Profit",axis="columns")
c
```

```python
d=b.Profit
d
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(c,d,test_size=0.9)
```

# VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns

a['Order Date'] = pd.to_datetime(a['Order Date'])
a['Month-Year'] = a['Order Date'].dt.to_period('M')


sales_trend = a.groupby('Month-Year')['Sales'].sum()


plt.figure(figsize=(12, 6))
sales_trend.plot(kind='bar', color='skyblue')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(a['Sales'], bins=30, color='green', alpha=0.7)
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.title('Sales Distribution Over Time')
plt.grid()
plt.show()
```

```python
a['Order Date'] = pd.to_datetime(a['Order Date'])  # Convert to datetime
plt.figure(figsize=(12, 6))
sns.boxplot(x=a['Order Date'].dt.month, y=a['Sales'])  # Use 'x' instead of 'a'
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Variation by Month')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=a, x='Sales', y='Profit', hue='Category', palette='coolwarm', alpha=0.7)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit by Category')
plt.grid()
plt.show()
```

```python
category_sales = a.groupby('Category')[['Sales', 'Profit']].sum()

plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('Category')
plt.ylabel('Total Value')
plt.title('Sales & Profit Distribution Across Categories')
plt.xticks(rotation=45)
plt.legend(['Sales', 'Profit'])
plt.grid()
plt.show()
```

```python
region_sales = a.groupby('Region')['Sales'].sum()

plt.figure(figsize=(10, 6))
region_sales.plot(kind='bar', color='lightcoral')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales Performance Across Regions')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
